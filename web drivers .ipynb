# Youtube Web Scrapping of Geeks for Geeks videos data :

```python
from selenium import webdriver
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.youtube.com/@GeeksforGeeksVideos/videos')
```

driver.page_source

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup.text.replace('\n','')
sp=soup.find_all('ytd-rich-item-renderer')
```

```python
sp[0].find_all('a')
```

```python
import pandas as pd
list1=[]
for sp in soup.find_all('ytd-rich-item-renderer'):
    title=sp.find_all('a')[3].text
    link=sp.find_all('a')[3].get('href')
    views=sp.find_all('span')[5].text
    time=sp.find_all('span')[6].text
    duration=sp.find_all('span')[0].text[3:7]
    img=sp.find('img').get('src')
    list1.append([title,link,views,time,duration,img])
    
    
    
```

```python
sp.find_all('span')
```

```python
df=pd.DataFrame(list1,columns=['title','link','views','time','duration','img'])
df.to_csv('youtube scrape.csv')
```

#  Gettiing the video details  :

```python
data1=[]
import time
import numpy as np
for i in df['link']:
    link='https://www.youtube.com/'+i
    driver=webdriver.Chrome()
    driver.get('https://www.youtube.com/'+i)
    time.sleep(2)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    try:
        title=soup.find('yt-formatted-string',class_='style-scope ytd-watch-metadata').text
    except:
        title=np.nan
    view=soup.find_all('span',class_='bold style-scope yt-formatted-string')
    try:
        views=view[0].text
    except:
         views=np,nan
    try:
        datetime=view[2].text
    except:
        datetime=np.nan
    try:
        likes=soup.find_all('span',class_='yt-core-attributed-string yt-core-attributed-string--white-space-no-wrap')[4].text
    except:
        likes=np.nan
    data1.append([title,views,datetime,likes])
```
